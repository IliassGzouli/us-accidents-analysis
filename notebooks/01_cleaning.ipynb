##  Analyse des valeurs manquantes

Nous analysons ici la quantité de valeurs manquantes dans chaque colonne
afin de définir une stratégie adaptée :
- suppression des colonnes trop vides,
- remplissage avec une valeur appropriée,
- interpolation (si applicable).


```python
df.isnull().sum()
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
cols_to_drop = ['End_Lat', 'End_Lng', 'Wind_Chill(F)', 'Weather_Timestamp',
    'Nautical_Twilight', 'Civil_Twilight', 'Sunrise_Sunset',
    'Astronomical_Twilight', 'Airport_Code', 'Zipcode']
df.drop(columns=cols_to_drop, inplace=True)
```

```python
cols_median = ['Wind_Speed(mph)', 'Visibility(mi)', 'Humidity(%)', 'Temperature(F)', 'Pressure(in)']
for col in cols_median:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cols_mode = ['Wind_Direction', 'Weather_Condition', 'Timezone']
for col in cols_mode:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df['Precipitation(in)'].fillna(0, inplace=True)
```

```python
cols_median = ['Wind_Speed(mph)', 'Visibility(mi)', 'Humidity(%)', 'Temperature(F)', 'Pressure(in)']
for col in cols_median:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cols_mode = ['Wind_Direction', 'Weather_Condition', 'Timezone']
for col in cols_mode:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df['Street'].fillna('Unknown', inplace=True)
```

```python
df['City'].fillna('Unknown', inplace=True)
```

```python
df['Description'].fillna('No description', inplace=True)
```
